```python
from matplotlib import pyplot as plt
import cv2
import numpy as np

img = cv2.imread('UM.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
height, width, channels = img.shape
print(img.shape)
```

```python
images=[]

for c in range(3):
    print(c)
    for i in [0.1,0.5,0.9]:
        temp_img = np.zeros((height,width,3), np.uint8)
        for h in range(height):
            for w in range(width):
                
                r,g,b=img[h,w]
                if c==0:
                    
                    temp_img[h,w]=int(r*i),g,b
                    
                elif c==1:
                    
                    temp_img[h,w]=r,int(g*i),b
                else:
                    
                    temp_img[h,w]=r,g,int(b*i)
        images.append(temp_img)
len(images)    

```

```python
plt.figure(figsize=(3,3)) 
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i])
    plt.axis('off')
plt.show()
```
